Приклад використання методу permutations() для створення списку перестановок також подається у комірці з кодом.

1) Створити 3 функції: функцію створення списку перестановок, функцію підрахунку добутків та функцію підрахунку загальної суми.

За бажання можна перевірити правильність роботи програми за допомогою функції np.linalg.det(a).

```
import numpy as np
import itertools

def random_matrix(dim):
    """
    The function generates dim x dim array of integers
    between 0 and 10.
    """
    matrix = np.random.randint(10, size = (dim, dim))
    return matrix
def permutations_list(n):
    """
    Ця функція робить список перестановок.
    """
    numb=0
    for i in range(1,n+1):
       numb=numb*10+i
    numb=str(numb)
    p_list = list(itertools.permutations(numb, n))
    return p_list
def product(p,permut_p,n):
    """
    Ця функція обчислює добуток.
    """
    n1 = len(permut_p)
    prod = [1]
    for i in range(0,n1-1):
        prod = prod+[1]
    for j in range(0,n1):
        for i in range(0,n):
            x = int(permut_p[0][i])
            y = int(permut_p[j][i])
            prod[j] = prod[j]*p[x-1][y-1]
    for e in range(0,n1):
        k = 0
        for j in range(0,n):
            for i in range(j,n):
                if permut_p[e][j]>permut_p[e][i]:
                    k = k + 1
        if(k%2!=0):
            prod[e] = prod[e]*(-1)
    return prod
def sum(ls):
    """
    Ця функція обчислює суму.
    """
    result = 0
    n1 = len(ls)
    for i in range(0,n1):
        result = result + ls[i]
    return result
try:
    a=int(input("Please enter size of square matrix"))
    if a>10 or a<1:
        raise ValueError
    else:
        a1=random_matrix(a)
        print(sum(product(a1,permutations_list(a),a)))
        print(np.linalg.det(a1))
except ValueError:
    print("Please enter number between 1 and 10")

```
